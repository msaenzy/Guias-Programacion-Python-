1) Invertir las columnas de una matriz

```python
import numpy as np
  
# matriz og
arr = np.array([
    
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]])
  
# invertir columna
flipped_arr = np.fliplr(arr)
  
print('Matriz original:\n', arr)
print('\n Matriz orden de columnas cambiado\n', flipped_arr)
```

Tabla de multiplicar de un nmero X usando el ciclo for.

```python
num = int(input ("Ingrese el numero que desee la tabla de multiplicar: "))      
  
print ("la tabla de multiplicar de: ", num)    
for count in range(1, 13):      
   print (num, 'x', count, '=', num * count)  
```

Resolver la serie 1 +  +  + 1/n usando el ciclo for .

```python
n=int(input("Enter the number of terms: "))
sum1=0
for i in range(1,n+1):
    sum1=sum1+(1/i)
    print(sum1)
print("The sum of series is",round(sum1,2))
```

Escribir una figura con nmeros dados un nmero x, usando el ciclo for. (Construir un
programa que, al recibir como dato un nmero entero positivo, escriba una figura como
la que se muestra a continuacin. Por ejemplo para N=6

```python
filas = 6

for i in range(filas):
    for j in range(i):
        
        print(i, end=' ')
    
    print('')
```

```python
filas = 5
for i in range(filas + 1, 0, -1):
    
    for j in range(0, i - 1):
    
        print("1", end=' ')
    print(" ")
```
