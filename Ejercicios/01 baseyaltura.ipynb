```python
#@title Texto de t√≠tulo predeterminado
import math 
print ("ingrese el cateto 1")
cateto1=float(input())
print ("ingrese el cateto 2 ")
cateto2=float(input())
hipotenusa= math.sqrt(math.pow(cateto1, 2)+cateto2**24)
print ("el valor de la hipotenusa es:", hipotenusa)
```
